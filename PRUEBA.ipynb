```python
#cv2 se utiliza para el reconocimiento de imagenes en python
import cv2
#se utiliza para la parte de arreglos
import numpy as np
#Os se utiliza para poder utilizar los directorios
import os
# matplotlib se utiliza para poder ver las graficas
import matplotlib.pyplot as plt

#Se crea una ruta
ruta_persona="/content/escritorio-persona/persona"
persona=[]
#Tamaño de imagen
img_size=150

#Se realiza un ciclo for para redimencionar al mismo tamaño todas las imagenes
for img in os.listdir(ruta_persona):
  img = cv2.imread(os.path.join(ruta_persona,img))
  #img_gray = cv2.cvtColor(img,cv2,COLOR_BGR2GRAY)
  img_resize = cv2.resize(img,(img_size,img_size))
  persona.append(img_resize)
  #Impresion de cuantos elementos tenemos
  print(len(persona))
  print(persona)
```

```python
persona = np.array(persona)
print(persona.shape)
#(Numero de imganes, tamaño de imagen, tamaño de imagen, canales de colores)
(15,150,150,3)
```

```python
print(persona[7].shape)
plt.figure()
#np squeeze se utiliza para reducir a 3 dimensiones
plt.imshow(np.squeeze(persona[7]))
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
# np concatenate junta las imagenes
images = np.concatenate([escritorio,persona])
#sirve para ver cuantas imagenes tenemos
print(len(images))
#lista de imagenes ya en arreglo
Images = np.array(images)
print(Images.shape)
#(numero de imagenes, tamaño de imagen, tamaño de imagen, canales de colores)
```

```python
#np repeat para reconocer las imagenes asignando valor de cero=0 y enseguida el numero de imagenes que tenemos
etiquetas_escritorio = np.repeat(0,15)
print(len(etiquetas_escritorio))
print(etiquetas_escritorio)
```

```python
#np repeat para reconocer las imagenes asignando valor de uno=1 y enseguida el numero de imagenes que tenemos
etiquetas_persona =np.repeat(1,15)
print(len(etiquetas_persona))
print(etiquetas_persona)
```

```python
# Nombres de las clases escritorio =0 y personas = 1
class_names=['escritorio','persona']
```

```python
#Se unen las etiquetas a traves de la funcion np concatenate en el orden que lo utilizamos
labels = np.concatenate ([etiquetas_escritorio,etiquetas_persona])
#ver cuantas etiquetas tenemos
print(len(labels))
print(labels)
#meterlas en un arreglo y mostrarlas
Labels = np.array(labels)
print(Labels.shape)
```

```python
#Graficar todas las imagenes
plt.figure(figsize=(10,10))
#rango de imagenes
for i in range(21):
  #el acomodo de las imagenes
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(Images[i])
  #, cmap=plt.cm.binary
  plt.xlabel(class_names[Labels[i]])
  plt.show()
```

```python
#Iniciar librerias de tensorflow
from __future__ import absolute_import, division, print_function, unicode_literals

variable_name = ""
try:
  # %tensorflow_version only exists in colab.
  %%!tensorflow_version 2.x
except Exception:
  pass

from __future__ import absolute_import, division, print_function, unicode_literals

 #Tensorflow and tf.keras
 #probar tensorflow en una consola de python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers, models
import tensorflow.keras.optimizers as Optimizer

  # Helper libraries
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

```python
#ENTRENAMIENTO
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(150, 150,3)),
    keras.layers.Dense(128, activation='relu'),

    keras.layers.Dense(2, activation='softmax'),
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
#numero de epocas
model.fit(Images, Labels, epochs=100)

trained=model.fit(Images, Labels, epochs=100)
```
